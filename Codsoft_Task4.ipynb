```python
# Remove leading/trailing spaces from column names (if any)
ratings.columns = ratings.columns.str.strip()

# If necessary, rename columns to match expected names
ratings = ratings.rename(columns={
    'userId': 'UserID',    # Adjust 'userId' to 'UserID' if needed
    'movieId': 'MovieID',  # Adjust 'movieId' to 'MovieID' if needed
    'rating': 'Rating'     # Adjust 'rating' to 'Rating' if needed
})
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the ratings data (ratings.csv)
ratings = pd.read_csv(r'C:\Users\sahup\ratings data\ratings.csv')  # Assuming 'ratings.csv' has UserID, MovieID, and Rating columns
movies = pd.read_csv(r'C:\Users\sahup\movies data\movies.csv')    # Assuming 'movies.csv' has the movie features including 'id' for MovieID

# Check the column names to ensure correct access
print("Column Names in Ratings Data:")
print(ratings.columns)

# Strip leading/trailing spaces from column names
ratings.columns = ratings.columns.str.strip()

# Rename columns if necessary to ensure they are as expected
ratings = ratings.rename(columns={
    'userId': 'UserID',    # Adjust 'userId' to 'UserID' if needed
    'movieId': 'MovieID',  # Adjust 'movieId' to 'MovieID' if needed
    'rating': 'Rating'     # Adjust 'rating' to 'Rating' if needed
})

# Check column names in the movies data
print("\nColumn Names in Movies Data:")
print(movies.columns)

# Strip leading/trailing spaces and rename columns if necessary
movies.columns = movies.columns.str.strip()
movies = movies.rename(columns={'id': 'MovieID'})  # Rename 'id' to 'MovieID'

# Display a sample of the movies data
print("\nSample of Movies Data:")
print(movies.head())

# Create the user-item matrix
user_movie_matrix = ratings.pivot(index='UserID', columns='MovieID', values='Rating')
user_movie_matrix = user_movie_matrix.fillna(0)

# Calculate the cosine similarity between users
user_similarity = cosine_similarity(user_movie_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

# Function to recommend movies to a user
def recommend_movies(user_id, user_movie_matrix, user_similarity_df, num_recommendations=5):
    user_ratings = user_movie_matrix.loc[user_id]  # Get the user's ratings
    
    # Get similarity scores for the user
    similar_users = user_similarity_df[user_id]  
    
    # Compute weighted ratings of all movies based on similar users' ratings
    weighted_ratings = user_movie_matrix.T.dot(similar_users) / np.abs(similar_users).sum()
    
    # Convert the weighted ratings to a Series for easier handling
    weighted_ratings = pd.Series(weighted_ratings, index=user_movie_matrix.columns)
    
    # Recommend the highest-rated movies that the user hasn't rated yet
    recommended_movies = weighted_ratings[~user_ratings.isna()].sort_values(ascending=False).head(num_recommendations)

    # Map MovieID to Movie Titles using the 'title' column in movies data
    recommended_movie_titles = movies[movies['MovieID'].isin(recommended_movies.index)]['title'].values

    return recommended_movie_titles

# Example: Recommend 5 movies for User 1
recommended_movies = recommend_movies(user_id=1, user_movie_matrix=user_movie_matrix, user_similarity_df=user_similarity_df, num_recommendations=5)
print("\nRecommended Movies for User 1:")
print(recommended_movies)

# Verify the changes
print("\nUpdated Column Names in Ratings Data:")
print(ratings.columns)

# Display a sample to ensure it looks correct
print("\nSample of Ratings Data:")
print(ratings.head())
```
